# Relación entre la Función Seno Hiperbólico y la Exponencial Compleja

## 1. Definiciones Fundamentales

### - Seno hiperbólico:
$$ \sinh(z) = \frac{e^z - e^{-z}}{2} $$

### - Exponencial compleja (Fórmula de Euler):
$$ e^{i\theta} = \cos\theta + i\sin\theta $$

---

## 2. Conexión con Números Complejos

### 2.1 Extensión al Plano Complejo
El seno hiperbólico puede extenderse a argumentos complejos usando su definición exponencial:

$$ \sinh(z) = \frac{e^{z} - e^{-z}}{2} \quad \text{para } z \in \mathbb{C} $$

### 2.2 Caso Particular: Argumento Imaginario Puro
Si tomamos $z = i\theta$ (donde $\theta \in \mathbb{R}$):

$$
\sinh(i\theta) = \frac{e^{i\theta} - e^{-i\theta}}{2}
$$

Usando la fórmula de Euler:
$$
= \frac{[\cos\theta + i\sin\theta] - [\cos(-\theta) + i\sin(-\theta)]}{2}
$$

Simplificando ($\cos$ es par, $\sin$ es impar):
$$
= \frac{[\cos\theta + i\sin\theta] - [\cos\theta - i\sin\theta]}{2} = i\sin\theta
$$

**Resultado clave**:
$$ \boxed{\sinh(i\theta) = i\sin\theta} $$

---

## 3. Relación Inversa

Podemos expresar funciones trigonométricas en términos de hiperbólicas para argumentos imaginarios:

$$
\sin(i\theta) = \frac{e^{i(i\theta)} - e^{-i(i\theta)}}{2i} = \frac{e^{-\theta} - e^{\theta}}{2i} = i\sinh\theta
$$

**Equivalencia**:
$$ \boxed{\sin(i\theta) = i\sinh\theta} $$

---

## 4. Forma General para Números Complejos

Para cualquier número complejo $z = x + iy$:

$$
\sinh(z) = \sinh(x + iy) = \sinh x \cos y + i\cosh x \sin y
$$

**Demostración**:
Usando la definición exponencial y propiedades de exponentes complejos:
$$
\sinh(x + iy) = \frac{e^{x + iy} - e^{-x - iy}}{2} = \frac{e^x (\cos y + i\sin y) - e^{-x} (\cos y - i\sin y)}{2}
$$

Agrupando términos reales e imaginarios:
$$
= \frac{(e^x - e^{-x})\cos y}{2} + i\frac{(e^x + e^{-x})\sin y}{2} = \sinh x \cos y + i\cosh x \sin y
$$

---

## 5. Representación en el Plano Complejo

| Componente          | Expresión                     |
|---------------------|-------------------------------|
| **Parte Real**      | $\sinh x \cos y$             |
| **Parte Imaginaria** | $\cosh x \sin y$             |

**Interpretación**:
- La parte real combina hiperbólicas (en $x$) con trigonométricas (en $y$)
- La parte imaginaria mezcla $\cosh$ y $\sin$ de forma similar

---

## 6. Aplicaciones Clave

1. **Mapeos conformes**: Transformaciones geométricas en el plano complejo
2. **Ecuaciones diferenciales**: Soluciones de ecuaciones en física matemática
3. **Relatividad**: Transformaciones de Lorentz con parámetros complejos
4. **Análisis de señales**: Procesamiento de señales complejas

---

## 7. Ejemplo Numérico

Sea $z = 1 + i\frac{\pi}{2}$:

$$
\sinh(1 + i\frac{\pi}{2}) = \sinh(1)\cos\left(\frac{\pi}{2}\right) + i\cosh(1)\sin\left(\frac{\pi}{2}\right)
$$

Calculando:
- $\sinh(1) \approx 1.175$, $\cosh(1) \approx 1.543$
- $\cos(\pi/2) = 0$, $\sin(\pi/2) = 1$

Resultado:
$$ \sinh(1 + i\frac{\pi}{2}) = 0 + i(1.543)(1) = i1.543 $$

---

# Funciones Trigonométricas Hiperbólicas: Explicación y Demostraciones

## 1. Introducción
Las funciones hiperbólicas son análogas a las funciones trigonométricas clásicas, pero están definidas usando la función exponencial. Surgen naturalmente en problemas relacionados con hipérbolas, termodinámica, teoría de relatividad y ecuaciones diferenciales.

---

## 2. Definiciones Básicas
Se definen a partir de combinaciones de exponenciales:

### - Seno hiperbólico (sinh):
$$ \sinh(x) = \frac{e^x - e^{-x}}{2} $$

### - Coseno hiperbólico (cosh):
$$ \cosh(x) = \frac{e^x + e^{-x}}{2} $$

### - Tangente hiperbólica (tanh):
$$ \tanh(x) = \frac{\sinh(x)}{\cosh(x)} = \frac{e^x - e^{-x}}{e^x + e^{-x}} $$

### Otras funciones:
- **Cotangente hiperbólica**: $\coth(x) = \frac{1}{\tanh(x)}$
- **Secante hiperbólica**: $\sech(x) = \frac{1}{\cosh(x)}$
- **Cosecante hiperbólica**: $\csch(x) = \frac{1}{\sinh(x)}$

---

## 3. Relación con la Hipérbola
Al igual que $(\cos t, \sin t)$ parametriza una circunferencia, $(\cosh t, \sinh t)$ parametriza la hipérbola $x^2 - y^2 = 1$.

**Demostración**:
Sustituimos las definiciones:
$$
\cosh^2(t) - \sinh^2(t) = \left(\frac{e^t + e^{-t}}{2}\right)^2 - \left(\frac{e^t - e^{-t}}{2}\right)^2
$$
Expandiendo:
$$
= \frac{e^{2t} + 2 + e^{-2t}}{4} - \frac{e^{2t} - 2 + e^{-2t}}{4} = \frac{4}{4} = 1
$$

---

## 4. Propiedades Clave

### - **Paridad**:
  - $\sinh(-x) = -\sinh(x)$ (impar).
  - $\cosh(-x) = \cosh(x)$ (par).

### - **Dominio y Rango**:
  - $\sinh(x)$: Dominio $\mathbb{R}$, Rango $\mathbb{R}$.
  - $\cosh(x)$: Dominio $\mathbb{R}$, Rango $[1, \infty)$.
  - $\tanh(x)$: Dominio $\mathbb{R}$, Rango $(-1, 1)$.

---

## 5. Identidades Importantes

### - **Identidad Fundamental**:
$$ \cosh^2(x) - \sinh^2(x) = 1 $$

### - **Fórmulas de Adición**:
$$
\sinh(a + b) = \sinh(a)\cosh(b) + \cosh(a)\sinh(b)
$$
$$
\cosh(a + b) = \cosh(a)\cosh(b) + \sinh(a)\sinh(b)
$$

### - **Doble Ángulo**:
$$ \sinh(2x) = 2\sinh(x)\cosh(x) $$
$$ \cosh(2x) = \cosh^2(x) + \sinh^2(x) $$

---

## 6. Derivadas e Integrales

### - **Derivadas**:
1. $\frac{d}{dx} \sinh(x) = \cosh(x)$  
   **Demostración**:
   $$ \frac{d}{dx} \left(\frac{e^x - e^{-x}}{2}\right) = \frac{e^x + e^{-x}}{2} = \cosh(x) $$

2. $\frac{d}{dx} \cosh(x) = \sinh(x)$  
3. $\frac{d}{dx} \tanh(x) = \sech^2(x)$

### - **Integrales**:
1. $\int \sinh(x) \, dx = \cosh(x) + C$  
2. $\int \cosh(x) \, dx = \sinh(x) + C$  
3. $\int \sech^2(x) \, dx = \tanh(x) + C$

---

## 7. Funciones Inversas

### - **Argumento del seno hiperbólico inverso**:
$$ \sinh^{-1}(x) = \ln\left(x + \sqrt{x^2 + 1}\right) $$

**Demostración**:
Sea $y = \sinh^{-1}(x) \implies x = \sinh(y) = \frac{e^y - e^{-y}}{2}$.  
Resolviendo para $y$, se obtiene la expresión logarítmica.

### - **Otras inversas**:
- $\cosh^{-1}(x) = \ln\left(x + \sqrt{x^2 - 1}\right) \quad (x \geq 1)$
- $\tanh^{-1}(x) = \frac{1}{2} \ln\left(\frac{1 + x}{1 - x}\right) \quad (|x| < 1)$

---

## 8. Aplicaciones
- **Catenaria**: La curva formada por una cadena colgante es $y = a \cosh(x/a)$.
- **Relatividad Especial**: Las transformaciones de Lorentz usan funciones hiperbólicas.
- **Cálculo Integral**: Aparecen en integrales de funciones racionales y ecuaciones diferenciales.


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración inicial
plt.rcParams.update({'font.size': 12})
x = np.linspace(-5, 5, 400)

# 1. Definir funciones
sinh = np.sinh(x)
cosh = np.cosh(x)

# 2. Crear figura con dos subtramas
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# --------------------------
# Gráfico 1: Función y derivada
# --------------------------
ax1.plot(x, sinh, label='sinh(x)', color='blue', lw=2)
ax1.plot(x, cosh, '--', label='cosh(x) (Derivada)', color='red', lw=2)
ax1.set_title('Función seno hiperbólico y su derivada')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Anotaciones explicativas
ax1.annotate('La pendiente en x=0 es 1\n(cosh(0) = 1)',
             xy=(0, 0), xytext=(1, 1),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

ax1.annotate('La derivada siempre crece\n(cosh(x) ≥ 1)',
             xy=(2, np.cosh(2)), xytext=(3, 4),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# --------------------------
# Gráfico 2: Demostración analítica
# --------------------------
ax2.axis('off')  # Ocultar ejes para el texto

# Texto con la demostración matemática
demostracion = (
    r"Demostración de $\frac{d}{dx}\sinh(x) = \cosh(x):$"
    "\n\n1. Definición:"
    r"$\sinh(x) = \frac{e^x - e^{-x}}{2}$"
    "\n\n2. Aplicamos derivada término a término:"
    r"$\frac{d}{dx}\left(\frac{e^x - e^{-x}}{2}\right)$"
    "\n\n3. Derivamos:"
    r"$\frac{1}{2}\left(\frac{d}{dx}e^x - \frac{d}{dx}e^{-x}\right)$"
    "\n\n4. Resultado intermedio:"
    r"$\frac{1}{2}\left(e^x - (-e^{-x})\right)$"
    "\n\n5. Simplificamos:"
    r"$\frac{e^x + e^{-x}}{2} = \cosh(x)$"
)

ax2.text(0.1, 0.5, demostracion, ha='left', va='center',
         fontsize=14, linespacing=1.8,
         bbox=dict(facecolor='white', alpha=0.9))

plt.tight_layout()
plt.show()
```
