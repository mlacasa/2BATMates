La integral de Riemann se puede entender como el límite de una suma de áreas bajo una curva a medida que aumentamos el número de particiones del intervalo. El método de exhaustión es la técnica que utiliza este enfoque para aproximar el valor de la integral. Aquí te explico el concepto paso a paso:

Conceptos Básicos
Función y Intervalo: Consideramos una función ( f(x) ) definida en un intervalo cerrado ([a, b]).

Particiones: Dividimos el intervalo ([a, b]) en ( n ) subintervalos de igual longitud. La longitud de cada subintervalo (denominada $ \Delta x $ )se puede calcular como:
$$
\Delta x = \frac{b - a}{n}
$$

**Suma de Riemann**:

Para cada subintervalo, seleccionamos un punto $ x_i^* $ dentro del subintervalo $[x_i, x_{i+1}]$ (puede ser el extremo izquierdo, derecho o cualquier punto en el subintervalo). La suma de Riemann se define como:
$$
S = \sum_{i=0}^{n-1} f(x_i^*) \cdot \Delta x
$$
Esta suma estima el área total bajo la curva mediante la suma de las áreas de rectángulos.

**Método de Exhausción**


El método de exhaustión se basa en el principio de que al aumentar el número de particiones ( n ), la suma de Riemann se vuelve una mejor aproximación del área real bajo la curva ( f(x) ). Aquí están los pasos clave:

Particionar el Intervalo: Empezamos con un número finito de particiones ( n ). Cuanto más alto sea ( n ), más precisos serán nuestros cálculos.

**Rectángulos Inferiores y Superiores:**

**Suma Inferior:** Utilizamos los valores de la función en los extremos izquierdos de cada subintervalo para calcular un área inferior. Esto significa que la suma considerará la altura más baja en cada rectángulo.


**Suma Superior:** Utilizamos los valores de la función en los extremos derechos para calcular un área superior, lo que implica que consideramos la altura más alta en cada rectángulo.


**Límite al Aumentar ( n ):** A medida que aumentamos ( n ) y, por lo tanto, disminuimos $ \Delta x $, ambas sumas inferior y superior se acercan al mismo valor, que es el área exacta bajo la curva entre ( a ) y ( b ). En el límite, cuando ( n ) tiende a infinito, la suma de Riemann converge a la integral definida:

$$
\int_a^b f(x) , dx = \lim_{n \to \infty} S_n
$$

Donde $ S_n $ es la suma de Riemann que se define como la suma inferior o superior, dependiendo de cómo seleccionemos los puntos dentro de los intervalos.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def area_superior_inferior(n): # En el ejemplo el intervalo es I = [0,1]
    a = 0
    b = 1
    delta_x = (b - a) / n   # En función de n el número de rectángulos


    suma_inferior = 0
    suma_superior = 0

    for i in range(n):
        x_i = a + i * delta_x
        suma_inferior += f(x_i) * delta_x

        x_i_plus_1 = a + (i + 1) * delta_x
        suma_superior += f(x_i_plus_1) * delta_x

    return suma_inferior, suma_superior

# Número de particiones
n = 20  # Puedes cambiar el valor de n para más precisión

area_inferior, area_superior = area_superior_inferior(n)

# Configuración del gráfico
x = np.linspace(0, 1, 100)
y = f(x)

# Graficar la función
plt.plot(x, y, label='f(x) = x^2', color='blue')

# Graficar las sumas inferior
for i in range(n):
    plt.fill_between([i/n, (i+1)/n], 0, f(i/n), color='lightgreen', alpha=0.5)

# Graficar las sumas superior
for i in range(n):
    plt.fill_between([(i)/n, (i+1)/n], 0, f((i+1)/n), color='lightcoral', alpha=0.5)

plt.title('Integral de Riemann de f(x) = x^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()

print(f"Área inferior: {area_inferior}")
print(f"Área superior: {area_superior}")
```

En el siguiente código lo que se grafica son los valores de las áreas superior e inferior, a medida que aumenta n, en el intervalo [0,100] y podemos ver cómo converge al valor de la integral, 1/3.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def area_superior_inferior(n):
    a = 0
    b = 1
    delta_x = (b - a) / n

    suma_inferior = 0
    suma_superior = 0

    for i in range(n):
        x_i = a + i * delta_x
        suma_inferior += f(x_i) * delta_x

        x_i_plus_1 = a + (i + 1) * delta_x
        suma_superior += f(x_i_plus_1) * delta_x

    return suma_inferior, suma_superior

# Valores de n en el intervalo [0, 100]
n_values = np.arange(1, 101)

# Calcular las áreas inferior y superior para cada valor de n
areas_inferiores = []
areas_superiores = []
for n in n_values:
  area_inferior, area_superior = area_superior_inferior(n)
  areas_inferiores.append(area_inferior)
  areas_superiores.append(area_superior)

# Graficar las áreas en función de n
plt.plot(n_values, areas_inferiores, label='Área Inferior')
plt.plot(n_values, areas_superiores, label='Área Superior')
plt.xlabel('Número de Particiones (n)')
plt.ylabel('Área')
plt.title('Áreas Inferior y Superior vs. Número de Particiones')
plt.legend()
plt.grid(True)
plt.show()
```

# CONDICIÓN DE INEGRABILIDAD DE UNA FUNCIÓN REAL DE VARIABLE REAL

Para que una función ( f ) sea integrable en un intervalo ([a, b]) según el criterio de Riemann, se deben cumplir las siguientes condiciones:

- **Continuidad**: Si la función es continua en el intervalo cerrado ([a, b]), entonces es integrable en ese intervalo.

**Discontinuidades Acotadas**: Si ( f ) tiene un número limitado de discontinuidades en ([a, b]), también será integrable.

El enfoque de las sumas inferior y superior se puede resumir de la siguiente manera:

- Suma Inferior ( L(f, P) ): Se calcula tomando el mínimo valor de ( f ) en cada subintervalo de una partición ( P ) de ([a, b]) y multiplicándolo por la longitud del subintervalo. La suma inferior es el límite inferior de la integral.

- Suma Superior ( U(f, P) ): Se calcula tomando el máximo valor de ( f ) en cada subintervalo de la misma partición ( P ) y multiplicándolo por la longitud del subintervalo. La suma superior es el límite superior de la integral.

Cuando el número de particiones ( n ) tiende a infinito, las siguientes condiciones deben cumplirse:

- **Convergencia**: La suma inferior ( L(f, P) ) y la suma superior ( U(f, P) ) deben converger a un mismo valor, que será el valor de la integral definida de la función en el intervalo ([a, b]). Esta es la condición de integrabilidad de Riemann.
En términos más formales, si:

$$
\lim_{||P|| \to 0} L(f, P) = \lim_{||P|| \to 0} U(f, P)
$$

donde ( ||P|| ) es el ancho máximo de los subintervalos de la partición, entonces la función ( f ) se considera integrable en ([a, b]).
